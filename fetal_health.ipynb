```python
X = df.drop(['fetal_health', 'histogram_median', 'histogram_min'], axis=1)
y = df[['fetal_health']]
```

```python
print("original")
print(y['fetal_health'].value_counts())
```

```python
from sklearn.preprocessing import StandardScaler

# Feature scaling - Z score normalization
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from imblearn.over_sampling import SMOTE

# Apply SMOTE (Synthetic Minority Over-sampling Technique)
smote = SMOTE(sampling_strategy='auto', random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_scaled, y)
```

```python
print("oversampled")
print(y_resampled['fetal_health'].value_counts())
```

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X_resampled,y_resampled,test_size=0.25,random_state=40)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_validate

# Define the models
models = [
    ("Random Forest", RandomForestClassifier(n_estimators=100)),
    ("K-Nearest Neighbors", KNeighborsClassifier(n_neighbors=5)),
    ("Gradient Boosting", GradientBoostingClassifier(n_estimators=100)),
]

# Loop through each model
for name, model in models:
    # Apply cross-validation
    cv_results = cross_validate(model, X_train, y_train, cv=5, scoring=['accuracy', 'precision_macro', 'recall_macro', 'f1_macro'])

    # Extract evaluation metrics
    accuracy = cv_results['test_accuracy'].mean()
    precision = cv_results['test_precision_macro'].mean()
    recall = cv_results['test_recall_macro'].mean()
    f1 = cv_results['test_f1_macro'].mean()

    # Print evaluation metrics
    print(f"Classifier: {name}")
    print(f"Accuracy: {accuracy}")
    print(f"Precision: {precision}")
    print(f"Recall: {recall}")
    print(f"F1-score: {f1}")
    print("\n")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc


num_classes = 3

# Binarize the labels
y_test_bin = label_binarize(y_test, classes=[1, 2, 3])  # Adjust classes here

# Iterate over the models list
for name, model in models:
    # Train the model
    model.fit(X_train, y_train)
    
    # Make predictions
    y_pred = model.predict(X_test)
    
    # Compute evaluation metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='macro')
    recall = recall_score(y_test, y_pred, average='macro')
    f1 = f1_score(y_test, y_pred, average='macro')
    
    # Compute the confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    
    # Compute ROC AUC curve
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    for i in range(num_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_pred == i+1)  # Adjust class labels here
        roc_auc[i] = auc(fpr[i], tpr[i])
    
    # Print evaluation metrics
    print(f"Classifier: {name}")
    print("Accuracy:", accuracy)
    print("Precision:", precision)
    print("Recall:", recall)
    print("F1-score:", f1)
    print("Confusion Matrix:")
    print(cm)
    print("\n")
    
    # Plot confusion matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['1', '2', '3'], yticklabels=['1', '2', '3'])
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title(f'Confusion Matrix - {name}')
    plt.show()
    
    # Plot ROC AUC curve
    plt.figure(figsize=(8, 6))
    for i in range(num_classes):
        plt.plot(fpr[i], tpr[i], label='Class {0} (AUC = {1:0.2f})'.format(i+1, roc_auc[i]))
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC AUC Curve - {name}')
    plt.legend(loc="lower right")
    plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from imblearn.over_sampling import SMOTE

# Assuming your data is in a DataFrame named 'data' and target column is 'fetal_health'
# Split features and target
X = df.drop('fetal_health', axis=1)
y = df['fetal_health']

# Print original class distribution
print("Original")
print(y.value_counts())

# Apply SMOTE to oversample the minority classes
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Print oversampled class distribution
print("Oversampled")
print(y_resampled.value_counts())

# Function to format labels with percentage and number
def format_autopct(pct, allvalues):
    absolute = int(pct/100.*sum(allvalues))
    return "{:.1f}% ({:d})".format(pct, absolute)

# Plot the pie charts vertically (up and down) with increased font size
fig, ax = plt.subplots(2, 1, figsize=(10, 14))

original_counts = y.value_counts()
ax[0].pie(original_counts, labels=original_counts.index, autopct=lambda pct: format_autopct(pct, original_counts), colors=plt.cm.Paired.colors, textprops={'fontsize': 17})
ax[0].set_title('Original Class Distribution', fontsize=17)

resampled_counts = y_resampled.value_counts()
ax[1].pie(resampled_counts, labels=resampled_counts.index, autopct=lambda pct: format_autopct(pct, resampled_counts), colors=plt.cm.Paired.colors, textprops={'fontsize': 17})
ax[1].set_title('Oversampled Class Distribution', fontsize=17)

plt.tight_layout()
plt.show()
```
